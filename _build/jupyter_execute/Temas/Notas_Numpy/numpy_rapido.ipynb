````python
print(a)
print(c[0,:,:])
r = a.dot(c[0,:,:])
print(r)
print(r.shape)
r.ndim
````

````python
ap = a.reshape((1,3))
print(ap)
ap.shape
rp = ap.dot(c[0,:,:])
print(rp)
print(rp.shape)
rp.ndim
````

````python
at = a.T
print(at)
at.shape
````

````python
apt = ap.T
print(apt)
apt.shape
````

````python
print(at)
r = np.ones((3,3)).dot(at)
print(r)
print(r.shape)
r.ndim
````

````python
print(apt)
r = np.ones((3,3)).dot(apt)
print(r)
print(r.shape)
r.ndim
````
